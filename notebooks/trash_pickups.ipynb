```python
missed.groupby("Trash Route").size().sort_values(ascending = False)
```

```python
missed["Trash Hauler"] = missed["Trash Hauler"].str.upper()
missed_by_hauler = missed["Trash Hauler"].value_counts()
```

```python
plt.figure()
missed_by_hauler.plot(kind = "bar")
plt.ylabel("Missed Pickups")
plt.show()
```

```python
counts_by_hauler = (missed.groupby(["Trash Hauler", "Incident Address"])).size().reset_index(name = "missed_count")
counts_by_hauler
```

```python
counts_by_hauler["fine"] = (counts_by_hauler["missed_count"] - 1) * 200
total_fines = counts_by_hauler.groupby("Trash Hauler")["fine"].sum()
total_fines
```
